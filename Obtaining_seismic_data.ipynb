```python
from obspy.clients.fdsn import Client
from obspy import UTCDateTime
```

```python
client="IRIS"
network="UU"
station="CTU"
Lococode="01"
Channel="HHZ"
starttime="2020-03-18T13:09:00.000"
endtime="2020-03-18T13:15:00.000"
t1=UTCDateTime(starttime)
t2=UTCDateTime(endtime)
client=Client(client)
st=client.get_waveforms(network,station,Lococode,Channel,t1,t2)
st.filter("bandpass", freqmin=0.1, freqmax=1)
st.plot()
```
